### Formal definition
A Petri net consist of a set S of species and a  set T of transitions together with  functions: $$i:S\times T\rightarrow N, o:S\times T\rightarrow N, $$ counting how many copies of each species shows as inputs and outputs of the transitions.


### Definition
A stochastic Petri net is a Petri net together with a function $$\tau:T\rightarrow (0,\infty)$$ giving a rate constant for each transition.


### Rate equation (classical harmonic oscilator)
How the expected number of things of each species changes with time.

### General equation
The ith species appears as input to a transition $m_i$ times and as output $n_i$ times, then 
$$\frac{dx_i}{dt} = r(n_i-m_i)x_1^{m_1}\cdots x_k^{m_k}$$
$x=(x_1^{m_1}\cdots x_k^{m_k}), m={m_1}\cdots {m_k}, n={n_1}\cdots {n_k}, x^m = x_1^{m_1}\cdots x_k^{m_k}$
$$\frac{dx}{dt} = r(n-m)x^m$$



If we have different transitions with different rate constants:
$$\frac{dx}{dt} = \sum_{\tau\in T} r(\tau)(n(\tau)-m(\tau))x^{m(\tau)}$$
